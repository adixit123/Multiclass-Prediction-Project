```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import make_pipeline
train=pd.read_csv("train.csv")
train=train.drop("id",axis=1)
test=pd.read_csv("test.csv")
test=test.drop("id",axis=1)
train
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split
train,pretest=train_test_split(train,test_size=0.15,stratify=train["NObeyesdad"],
                               random_state=42)
validation=train["NObeyesdad"]
```

```python
pretest
```

```python
from sklearn.preprocessing import OrdinalEncoder
oencoder=OrdinalEncoder()
train_char=train.select_dtypes(include=[object]).drop("NObeyesdad", axis=1)
test_char=test.select_dtypes(include=[object])
pretest_char=pretest.select_dtypes(include=[object])
```

```python
train_chararr=oencoder.fit_transform(train_char)
test_chararr=oencoder.fit_transform(test_char)
pretest_chararr=oencoder.fit_transform(pretest_char)
```

```python
train_char=pd.DataFrame(train_chararr,columns=train_char.columns,
                        index=train_char.index)
test_char=pd.DataFrame(test_chararr,columns=test_char.columns,
                        index=test_char.index)
pretest_char=pd.DataFrame(pretest_chararr,columns=pretest_char.columns,
                        index=pretest_char.index)
```

```python
from sklearn.preprocessing import StandardScaler
train_num=train.select_dtypes(include=[np.number])
test_num=test.select_dtypes(include=[np.number])
pretest_num=pretest.select_dtypes(include=[np.number])
sscaler=StandardScaler()
```

```python
train_numarr=sscaler.fit_transform(train_num)
test_numarr=sscaler.fit_transform(test_num)
pretest_numarr=sscaler.fit_transform(pretest_num)
train_num=pd.DataFrame(train_numarr,columns=train_num.columns,index=train_num.index)
test_num=pd.DataFrame(test_numarr,columns=test_num.columns,index=test_num.index)
pretest_num=pd.DataFrame(pretest_numarr,columns=pretest_num.columns,index=pretest_num.index)
X=pd.concat([train_num,train_char],axis=1)
Y2=pd.concat([test_num,test_char],axis=1)
Y1=pd.concat([pretest_num,pretest_char],axis=1)
```

```python
clf=GradientBoostingClassifier(n_estimators=100,learning_rate=1.0,
                              max_depth=1,random_state=0)
X
```

```python
clf.fit(X,validation.values.ravel())

```

```python
pretest
```

```python
Y1ev=pretest["NObeyesdad"]
Y1=Y1.drop("NObeyesdad",axis=1)
preresults=clf.predict(Y1)
```

```python
preresults
preresults=pd.DataFrame(preresults,columns=["NObeyesdad"])
preresults
```

```python
from sklearn.metrics import accuracy_score
clf=GradientBoostingClassifier(n_estimators=100,learning_rate=0.75,max_depth=1,random_state=42)
clf.fit(X,validation)
preresults=clf.predict(Y1)
accuracy_score(preresults,Y1ev)
```

```python
results=clf.predict(Y2)
final=pd.DataFrame(results,columns=["NObeyesdad"])
id1=pd.read_csv("test.csv")
id1=pd.DataFrame(id1["id"],columns=["id"])
finalres=pd.concat([id1,final],axis=1)
```
